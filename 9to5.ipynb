1. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">Write a query that returns the series_id, industry_code, industry_name, and value from the january_2017 table but only if that value is greater than the average value for annual_2016 of data_type_code 82.</span>

````sql
SELECT * FROM series
JOIN datatype
ON series.data_type_code = datatype.data_type_code
WHERE series.data_type_code = 82
````

````sql
/*First I join the tables: january_2017, series, and industry to display the following
  coloumns: series_id, industry_code, industry_name, and value
  
  Next I set a condition where the value must be larger than the AVERAGE 2016 values 
  where the datacode for the series is 82 (AGGREGATE WEEKLY PAYROLLS OF PRODUCTION AND NONSUPERVISORY EMPLOYEES)*/

SELECT s.series_id, s.series_title, i.industry_code, i.industry_name, j.[value]
FROM january_2017 AS j
JOIN series AS s
ON j.series_id = s.series_id
JOIN industry as i
ON i.industry_code = s.industry_code
WHERE s.data_type_code = 82 and j.[value] >
    (SELECT AVG(a.value)
     FROM annual_2016 AS a
     WHERE a.series_id IN
        (SELECT series_id
        FROM series
        WHERE data_type_code = 82))
````

2. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">Create a Union table comparing average weekly earnings&nbsp; of production and nonsupervisory employees between annual_16 and january_17 using the data type 30.&nbsp; Round to the nearest penny.&nbsp; You should have a column for the average earnings and a column for the year, and the period.</span>

````sql
/*First I created a union between the annual2016 table and the january2017 table
  Then I set a condtion on the series id with both codtions (data type is 30 and data name describes
   'average weekly earning of production and nonsupervisory employees')*/

SELECT a.[year], a.[period], AVG(a.[value])  AS 'AVG OF Average weekly earnings of production and nonsupervisory employees'
FROM annual_2016 AS a
WHERE a.series_id IN
    (SELECT series_id
    FROM series
    WHERE series_title LIKE 'Average weekly earnings of production and nonsupervisory employees'
            AND data_type_code = 30)
GROUP BY a.[year], a.[period]

UNION

SELECT j.[year], j.[period], AVG(j.[value]) AS 'AVG OF Average weekly earnings of production and nonsupervisory employees'
FROM january_2017 AS j
WHERE j.series_id IN
    (SELECT series_id
    FROM series
    WHERE series_title LIKE 'Average weekly earnings of production and nonsupervisory employees'
            AND data_type_code = 30)
GROUP BY j.[year], j.[period]
````

IN January 2017 the **<u>Reinsurance Carriers</u>** industry made the most AVERAGE weekly earnings out of all production and nonsupervisory employees with $**1,810.59**

````sql
/*QUERY finding the AVG of Average weekly earnings of production and nonsupervisory employees grouped by industry name and year(period)*/
SELECT a.[year], i.industry_name, AVG(a.[value])  AS 'AVG OF Average weekly earnings of production and nonsupervisory employees'
FROM annual_2016 AS a
JOIN series AS s
ON a.series_id = s.series_id
JOIN industry AS i
ON s.industry_code = i.industry_code
WHERE a.series_id IN
    (SELECT series_id
    FROM series
    WHERE series_title = 'Average weekly earnings of production and nonsupervisory employees'
            AND data_type_code = 30)
GROUP BY a.[year], i.industry_name


UNION

SELECT j.[year], i.industry_name, AVG(j.[value]) AS 'AVG OF Average weekly earnings of production and nonsupervisory employees'
FROM january_2017 AS j
JOIN series AS s
ON j.series_id = s.series_id
JOIN industry AS i
ON s.industry_code = i.industry_code
WHERE j.series_id IN
    (SELECT series_id
    FROM series
    WHERE series_title = 'Average weekly earnings of production and nonsupervisory employees'
            AND data_type_code = 30)
GROUP BY j.[year], i.industry_name
ORDER BY 'AVG OF Average weekly earnings of production and nonsupervisory employees' DESC
````
